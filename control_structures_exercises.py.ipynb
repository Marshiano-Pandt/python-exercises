```python
'''break and continue

Write a program that prompts the user for a 
positive integer. Next write a loop that prints
out the numbers from the number the user entered
down to 1.'''

loop_control = True

while loop_control == True:
    num = int(input("Enter a positve number "))
    
    if num > 0:
        break
    else:
        print("Enter a positive number. Try again ")
    
for i in range (num,0,-1):
    print(i)
    
```

```python
# 5. Convert given number grades into letter grades.

"""Prompt the user for a numerical grade from 0 to 
100

Display the corresponding letter grade

Prompt the user to continue

Assume that the user will enter valid integers for
the grades

The application should only continue if the user 
agrees to
Grade Ranges:

A : 100 - 88
B : 87 - 80
C : 79 - 67
D : 66 - 60
F : 59 - 0"""

loop_control = True

while loop_control == True:
    num_grade = int(input("Enter a numerical grade (0-100): "))
    
    if 88 <= num_grade <= 100:
        letter_grade = "A"
    elif 80 <= num_grade <= 87:
        letter_grade = "B"
    elif 67 <= num_grade <= 79:
        letter_grade = "C"
    elif 60 <= num_grade <= 66:
        letter_grade = "D"
    elif 0 <= num_grade <= 59:
        letter_grade = "F"
    else:
        print("Invalid input. Please enter a numerical grade between 0 and 100. ")
        continue
        
    print(f"The letter grade for {num_grade} is {letter_grade}")
    
    user_response = input("Do you want to continue? (yes/no): ").lower()
    if user_response != 'yes':
        break

print("Goodbye!")


```
